```python
# Load necessary packages
import xarray as xr
```

```python
data_dir = '/Volumes/Iris 300 GB/CMIP6'
```

```python
#Function to import cmip6 data
# For 'sce' choose ['historical',  'piControl', 'ssp119', 'ssp126', 'ssp245', 'ssp370', 'ssp585']
# For 'var' choose ['zos',  'vas', 'uas', 'tos', 'ps'] which are respectively 
# sea level height, meridional wind, zonal wind, sea surface temperature, pressure at sea surface

def import_cmip6(sce, var):
    if sce == 'piControl':
        return xr.open_dataset('/Volumes/Iris 300 GB/CMIP6/cmip6_uas_piControl/cmip6_uas_piControl_ACCESS-CM2_950_1449.nc')
    else: 
        return xr.open_mfdataset(f'{data_dir}/cmip6_{var}_{sce}/cmip6_{var}_{sce}_*.nc')

# Function to import all different data series
def import_cmip6_all(var):
    hist = import_cmip6('historical', var)
    piCon = import_cmip6('piControl', var)
    ssp119 = import_cmip6('ssp119', var)
    ssp126 = import_cmip6('ssp126', var)
    ssp245 = import_cmip6('ssp245', var)
    ssp370 = import_cmip6('ssp370', var)
    ssp585 = import_cmip6('ssp585', var)
    return hist, piCon, ssp119, ssp126, ssp245, ssp370, ssp585
```

```python
# Function from https://github.com/dlebars/CMIP_SeaLevel/blob/master/notebooks/PlotZOS.ipynb

def plot_all_models2(ds):
    '''Use faceting feature from xarray'''
    cbar_kwargs = {'orientation': 'horizontal','label': 'cm', 'pad': 0.05}
    ds.CorrectedReggrided_zos.isel(time=-1).plot(x='lon', y='lat', col='model', 
                                                    col_wrap=4, robust=True, 
                                                    cbar_kwargs=cbar_kwargs)
```
