### Notebook to examine the inverse barometer effect


Similar method as used by Le Bars: (https://github.com/dlebars/SLBudget/blob/master/code/SeaLevelContrib.py)


```python
# Import necessary functions
import xarray as xr
import pandas as pd
import matplotlib.pyplot as plt


# Declare path
path = '/Users/iriskeizer/Projects/ClimatePhysics/Thesis/Data/observations/Pressure/'
path_coords = '/Users/iriskeizer/Projects/ClimatePhysics/Thesis/Data/observations/Coordinates/'
path_fig = '/Users/iriskeizer/Projects/ClimatePhysics/Thesis/Figures/Data examination/IBE/'
```

```python
# Import data
pres_era5 = xr.open_dataset(path+'pres_annual_era5.nc')
pres_20cr = xr.open_dataset(path+'pres_annual_20cr.nc')
```

```python
# Import tide gauge station coordinates
tg_coords = pd.read_csv(path_coords+'tgstations.csv', index_col='station')
tg_coords = tg_coords
```

#### Compute IBE at tide gauge locations

```python
def ibe_df(pressure):
    
    # Declare variables
    rho = 1030 # Density of ocean water
    g = 9.81 # Acceleration of gravity on Earth's surface
    
    
    df = pd.DataFrame({'time':pressure.year.values})
    df = df.set_index('time')
    
    for station in tg_coords.index:
        pres = pressure.sel(lat = tg_coords['lat'][station], lon = tg_coords['lon'][station], method='Nearest').pressure.values
        pres = pres - pres.mean() # Calculate anomaly while i have no idea why
        
        ibe = -pres/(rho*g)*100 # Inverse barometer effect in cm
        df[station] = ibe - ibe.mean() # Calculate anomaly?? again and save in DataFrame
        
    df['Average'] = df.mean(axis=1)
    
    return df
```

```python
ibe_era5 = ibe_df(pres_era5)
ibe_20cr = ibe_df(pres_20cr)
```

But how to remove this inverse barometer effect from the tide gauge data since time series are different? That does not matter since for the regression only the overlapping time series are used so I can just remove the ibe before regressing

```python
def remove_ibe(tg_data, ibe):
    
    # Create dataframes of equal time span
    tg_data = tg_data[tg_data.index.isin(ibe.index)]
    ibe = ibe[ibe.index.isin(tg_data.index)]
    
    df = pd.DataFrame({'time':tg_data.index.values})
    df = df.set_index('time')
    
    
    for station in tg_data.columns:
        removed = tg_data[station].values - ibe[station].values
        df[station] = removed
        
    return df
```

```python
# Remove ibe from tide gauge data
tg_wh_ibe_era5 = remove_ibe(tg_data_df, ibe_era5)
tg_wh_ibe_20cr = remove_ibe(tg_data_df, ibe_20cr)
```

```python
# Import the data 

# Pressure
path = '/Users/iriskeizer/Projects/ClimatePhysics/Thesis/Data/cmip6/'
ps = xr.open_dataset(path + f'Pressure/pressure_annual_historical.nc') 


#SLH
path = f'/Users/iriskeizer/Projects/ClimatePhysics/Thesis/Data/cmip6/SLH/slh_annual_historical.nc'
zos = xr.open_dataset(path) 
```

```python
# Only use models occuring in both datasets and in the other wind models
models = ['ACCESS-CM2', 'ACCESS-ESM1-5', 'BCC-CSM2-MR', 'BCC-ESM1',
       'CAMS-CSM1-0', 'CAS-ESM2-0', 'CMCC-CM2-SR5', 'CMCC-ESM2',
       'CNRM-CM6-1', 'CNRM-ESM2-1', 'CanESM5', 'CanESM5-CanOE',
       'EC-Earth3', 'EC-Earth3-AerChem', 'EC-Earth3-CC', 'EC-Earth3-Veg',
       'EC-Earth3-Veg-LR', 'FGOALS-f3-L', 'GFDL-CM4', 'GFDL-ESM4',
       'GISS-E2-1-G', 'GISS-E2-1-H', 'HadGEM3-GC31-LL', 'HadGEM3-GC31-MM',
       'INM-CM4-8', 'INM-CM5-0', 'IPSL-CM6A-LR', 'MIROC-ES2L', 'MIROC6',
       'MPI-ESM-1-2-HAM', 'MPI-ESM1-2-HR', 'MPI-ESM1-2-LR', 'MRI-ESM2-0',
       'NESM3', 'NorCPM1', 'UKESM1-0-LL']
zos = zos.where(zos.model.isin(models), drop=True)
ps = ps.where(ps.model.isin(models), drop=True)
```

```python
# Import cmip6 station coordinates
path_coords = '/Users/iriskeizer/Projects/ClimatePhysics/Thesis/Data/cmip6/Coordinates/'
zos_coords = pd.read_csv(path_coords+'hist.csv', index_col='station')
```

```python
# select pressure near tide gauge station
ps_lst = []
for index, row in zos_coords.iterrows():
    ps_lst.append(ps.sel(lon = row.lon, lat = row.lat, method = 'nearest'))
    ps_lst[-1] = ps_lst[-1].drop(['lat', 'lon'])
    
    
# Create a dataarray
ps = xr.concat(ps_lst, zos.station.values[:-1]).rename({'concat_dim':'station'})
```

```python
# Declare variables
rho = 1030 # Density of ocean water
g = 9.81 # Acceleration of gravity on Earth's surface

# Obtain anomalies of pressure
ps = ps.ps - ps.ps.mean(dim='time')

# Calculate ibe
ibe = -ps/(rho*g)*100
ibe = ibe - ibe.mean(dim='time') # Anomalies of ibe

# Remove ibe
zos_wh_ibe = zos.zos - ibe # zos minus the ibe 

# Obtain average station
average = zos_wh_ibe.mean('station')
average = average.assign_coords({'station':'Average'})

# Concat to original dataarray
zos_wh_ibe = xr.concat([zos_wh_ibe, average], dim='station')
```

```python
# Plot total zos and with ibe removed for average station
station = 'Average'


models = zos.model.values

n_col = 4
n_row = 9

y_min = 15
y_max = -15
    
fig, axs = plt.subplots(n_row, n_col, figsize=(24, 20))

for i in range(n_row):
    for j in range(n_col):
        
        ax = axs[i,j]
        
        ax.plot(zos.time.values, zos.zos.sel(station=station, model=models[4*i+j]), label = 'Zos')
        ax.plot(zos_wh_ibe.time.values, zos_wh_ibe.sel(station=station, model=models[4*i+j]), label = 'Zos without ibe')
        
        ax.set_xlabel('time [y]')
        if j == 0:
            ax.set_ylabel('zos [cm]')
        if i == 0 and j == 0:
            ax.legend()
        ax.axhline(color='k', linestyle='--', linewidth = 1)
        ax.set_title('model = ' + models[4*i+j])
        ax.set_ylim(y_min,y_max)
        plt.tight_layout()
        

plt.savefig(path_fig+'zos_wh_ibe.png')
```
