```python
# Import necessary functions
import xarray as xr
import pandas as pd
import numpy as np


# Declare path
path = '/Users/iriskeizer/Projects/ClimatePhysics/Thesis/Data/observations/Pressure/'
path_coords = '/Users/iriskeizer/Projects/ClimatePhysics/Thesis/Data/observations/Coordinates/'
path_fig = '/Users/iriskeizer/Projects/ClimatePhysics/Thesis/Figures/Data examination/Nodal/'
```

```python
# Import tide gauge station coordinates
tg_coords = pd.read_csv(path_coords+'tgstations.csv', index_col='station')
tg_coords = tg_coords[:-1] # Remove average station
```

```python
# Import tide gauge data
path = '/Users/iriskeizer/Projects/ClimatePhysics/Thesis/Data/observations/SLH/tg_data.csv'
tg_data_df = pd.read_csv(path)
tg_data_df = tg_data_df.set_index('time').drop('Average', axis=1)
tg_data_df.plot(figsize=(9,3), ylabel = 'SLH [cm]', xlabel = 'time [y]')
```

#### Compute the nodal potential at tide gauge locations

```python
def nodal_tides_potential(lat, time_years):
    """
    Function to obtain the nodal cycle
    
    """
    h2 = 0.6032
    k2 = 0.298

    #nodal cycle correction
    A = 0.44*(1+k2-h2)*20*(3*np.sin(lat*np.pi/180.)**2-1)/10  # mm to cm
    nodcyc = A*np.cos((2*np.pi*(time_years-1922.7))/18.61 + np.pi)
    
    return nodcyc.values
```

```python
nodal = pd.DataFrame({'time':tg_data_df.index.values})
nodal = nodal.set_index('time')

for station in tg_coords.index:

        #Remove nodal cycle
        nodal_cor = nodal_tides_potential(tg_coords['lat'][station], tg_data_df.index)
        
        # Add to nodal dataframe
        nodal[station] = nodal_cor
        
tg_wh_nodal = tg_data_df - nodal
```

```python
nodal.plot(figsize = (9,3), xlabel='time [y]', ylabel='Nodal potential [cm]')
plt.savefig(path_fig+'nodal_at_tg_stations')
```

```python
# Obtain averages
tg_data_df['Average'] = tg_data_df.mean(axis=1)
tg_wh_nodal['Average'] = tg_wh_nodal.mean(axis=1)
```

```python
import matplotlib.pyplot as plt

for station in tg_data_df.columns:
    plt.figure(figsize=(9,3))
    plt.plot(tg_data_df.index, tg_data_df[station], label = 'Original tide gauge data')
    plt.plot(tg_wh_nodal.index, tg_wh_nodal[station], label = 'Tide gauge data with nodal cycle removed')
    plt.xlim(1889, 2021)
    plt.ylim(-25,25)
    plt.title(f'station = {station}')
    plt.xlabel('time [y]')
    plt.ylabel('SLH [cm]')
    plt.legend()
    plt.tight_layout()
    plt.savefig(path_fig+f'tg_wh_nodal_{station}.png')
```

#### The nodal cycle for model data

```python
# Import the slh data 
path = f'/Users/iriskeizer/Projects/ClimatePhysics/Thesis/Data/cmip6/SLH/slh_annual_historical.nc'
zos = xr.open_dataset(path) 
```

```python
# Only use models occuring in both datasets and in the other wind models
models = ['ACCESS-CM2', 'ACCESS-ESM1-5', 'BCC-CSM2-MR', 'BCC-ESM1',
       'CAMS-CSM1-0', 'CAS-ESM2-0', 'CMCC-CM2-SR5', 'CMCC-ESM2',
       'CNRM-CM6-1', 'CNRM-ESM2-1', 'CanESM5', 'CanESM5-CanOE',
       'EC-Earth3', 'EC-Earth3-AerChem', 'EC-Earth3-CC', 'EC-Earth3-Veg',
       'EC-Earth3-Veg-LR', 'FGOALS-f3-L', 'GFDL-CM4', 'GFDL-ESM4',
       'GISS-E2-1-G', 'GISS-E2-1-H', 'HadGEM3-GC31-LL', 'HadGEM3-GC31-MM',
       'INM-CM4-8', 'INM-CM5-0', 'IPSL-CM6A-LR', 'MIROC-ES2L', 'MIROC6',
       'MPI-ESM-1-2-HAM', 'MPI-ESM1-2-HR', 'MPI-ESM1-2-LR', 'MRI-ESM2-0',
       'NESM3', 'NorCPM1', 'UKESM1-0-LL']
zos = zos.where(zos.model.isin(models), drop=True)
```

```python
# Import cmip6 station coordinates
path_coords = '/Users/iriskeizer/Projects/ClimatePhysics/Thesis/Data/cmip6/Coordinates/'
zos_coords = pd.read_csv(path_coords+'hist.csv', index_col='station')
```

#### Compute the nodal potential at cmip6 locations

```python
def nodal_tides_potential(lat, time_years):
    """
    Function to obtain the nodal cycle
    
    """
    h2 = 0.6032
    k2 = 0.298

    #nodal cycle correction
    A = 0.44*(1+k2-h2)*20*(3*np.sin(lat*np.pi/180.)**2-1)/10  # mm to cm
    nodcyc = A*np.cos((2*np.pi*(time_years-1922.7))/18.61 + np.pi)
    
    return nodcyc
```

```python
nodal_cmip6 = pd.DataFrame({'time':zos.time.values})
nodal_cmip6 = nodal_cmip6.set_index('time')

zos_stat_lst = []
for station in zos_coords.index:

        # Obtain nodal potential
        nodal_cor = nodal_tides_potential(zos_coords['lat'][station], zos.time.values)
        
        # Add to nodal dataframe
        nodal_cmip6[station] = nodal_cor
        
        # Remove nodal cycle
        zos_mod_lst = []
        for model in models:
            wh_nodal = zos.sel(station=station, model=model) - nodal_cor
            zos_mod_lst.append(wh_nodal.drop(['station', 'model']))
        zos_stat_lst.append(xr.concat(zos_mod_lst, models).rename({'concat_dim':'model'}))
        
zos_wh_nodal = xr.concat(zos_stat_lst, zos.station.values[:-1]).rename({'concat_dim':'station'})


# Obtain average station
average = zos_wh_nodal.mean('station')
average = average.assign_coords({'station':'Average'})

# Concat to original dataarray
zos_wh_nodal = xr.concat([zos_wh_nodal, average], dim='station')
```

```python
# Plot total zos and with nodal removed for average station
station = 'Average'


n_col = 4
n_row = 9

y_min = 15
y_max = -15
    
fig, axs = plt.subplots(n_row, n_col, figsize=(24, 20))

for i in range(n_row):
    for j in range(n_col):
        
        ax = axs[i,j]
        
        ax.plot(zos.time.values, zos.zos.sel(station=station, model=models[4*i+j]), label = 'Zos')
        ax.plot(zos_wh_nodal.time.values, zos_wh_nodal.zos.sel(station=station, model=models[4*i+j]), label = 'Zos without nodal')
        
        ax.set_xlabel('time [y]')
        if j == 0:
            ax.set_ylabel('zos [cm]')
        if j == 0:
            ax.legend()
        ax.axhline(color='k', linestyle='--', linewidth = 1)
        ax.set_title('model = ' + models[4*i+j])
        ax.set_ylim(y_min,y_max)
        plt.tight_layout()
        
plt.savefig(path_fig+'zos_wh_nodal.png')
```
