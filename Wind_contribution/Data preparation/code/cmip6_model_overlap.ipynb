### Notebook to check which models are available for all variables and scenarios

```python
import xarray as xr
```

```python
# For each variable least models are available for ssp_126
```

```python
zos = xr.open_dataset('/Users/iriskeizer/Projects/ClimatePhysics/Thesis/Data/cmip6/SLH/slh_annual_ssp126.nc') 
uas = xr.open_dataset('/Users/iriskeizer/Projects/ClimatePhysics/Thesis/Data/cmip6/Wind/wind_annual_ssp126.nc') 
ps = xr.open_dataset('/Users/iriskeizer/Projects/ClimatePhysics/Thesis/Data/cmip6/Pressure/pressure_annual_ssp126.nc') 
tos = xr.open_dataset('/Users/iriskeizer/Projects/ClimatePhysics/Thesis/Data/cmip6/SST/tos_annual_ssp126.nc') 
```

```python
models = zos.where(zos.model.isin(uas.model), drop = True).model
models = models.where(models.model.isin(ps.model), drop = True).model
models = models.where(models.model.isin(tos.model), drop = True).model
```

```python
zos_time = zos.where(zos.model.isin(models), drop = True)
```
